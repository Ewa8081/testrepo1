```python
# Import necessary libraries
import pandas as pd

# Load the Stack Overflow survey data
dataset_url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"
df = pd.read_csv(dataset_url)

# Display the first few rows
print(df.head())
```

```python
#Summarize the dataset by displaying the column data types, counts, and missing values.¶
# Step 1: Display column data types and non-null counts
print("\nBasic Info:")
print(df.info())

# Step 2: Count missing values per column
missing_values = df.isnull().sum()

# Step 3: Create a summary table
summary_df = pd.DataFrame({
    'Data Type': df.dtypes,
    'Non-Null Count': df.count(),
    'Missing Values': missing_values
})

print("\nDataset Summary:")
print(summary_df)
```

```python
#Generate basic statistics for numerical columns.
# Generate basic statistics for numerical columns
stats = df.describe()

print("\nBasic Statistics for Numerical Columns:")
print(stats)
```

```python
#Identify inconsistent or irrelevant entries in specific columns (e.g., Country).
# Step 1: View unique values in the Country column
print("Unique Country Entries:")
print(df['Country'].unique())

# Step 2: Identify inconsistent formats (e.g., different spellings of the same country)
df['Country'] = df['Country'].str.strip().str.lower()  # Normalize case and spaces

# Step 3: Standardize known country names
corrections = {
    'usa': 'United States',
    'united states': 'United States',
    'u.s.': 'United States',
    'canada': 'Canada',
    'germany': 'Germany',
    'germany ': 'Germany'  # Handling potential trailing spaces
}

df['Country'] = df['Country'].replace(corrections)

# Step 4: Identify irrelevant values (e.g., 'N/A', 'Unknown')
irrelevant_values = ['n/a', 'unknown', 'not specified', 'null', 'other']
df['Country'] = df['Country'].apply(lambda x: None if x in irrelevant_values else x)

# Step 5: Display cleaned values
print("\nCleaned Country Entries:")
print(df['Country'].unique())
```

```python
#Standardize entries in columns like Country or EdLevel by mapping inconsistent values to a consistent format.
df['Country'] = df['Country'].str.strip().str.lower()

# Step 2: Define a mapping dictionary for standardization
country_mapping = {
    'usa': 'United States',
    'united states': 'United States',
    'u.s.': 'United States',
    'canada': 'Canada',
    'germany': 'Germany',
    'germany ': 'Germany',  # Handling potential extra spaces
    'france': 'France'
}

# Step 3: Apply the mapping
df['Country'] = df['Country'].replace(country_mapping)

# Step 4: Handle missing or irrelevant values
irrelevant_values = ['n/a', 'unknown', 'not specified', 'null', 'other']
df['Country'] = df['Country'].apply(lambda x: None if x in irrelevant_values else x)

# Step 5: Display cleaned values
print("\nCleaned Country Entries:")
print(df['Country'].unique())

# Optional: Display cleaned DataFrame
print("\nUpdated DataFrame:")
print(df)
```

```python
#Encode the Employment column using one-hot encoding.
# Step 1: Handle missing values (fill with 'Unknown')
df['Employment'].fillna('Unknown', inplace=True)

# Step 2: Apply one-hot encoding
df_encoded = pd.get_dummies(df, columns=['Employment'], prefix='Emp')

# Step 3: Display the transformed DataFrame
print("\nOne-Hot Encoded DataFrame:")
print(df_encoded)
```

```python
# Identify columns with the highest number of missing values.
# Count missing values per column
missing_values = df.isnull().sum()

# Sort columns by highest missing values
missing_values_sorted = missing_values.sort_values(ascending=False)

# Display the results
print("\nColumns with Missing Values (Sorted):")
print(missing_values_sorted)
```

```python
# Impute missing values in numerical columns (e.g., `ConvertedCompYearly`) with the mean or median.
# Step 1: Check missing values before imputation
print("\nMissing Values Before Imputation:")
print(df.isnull().sum())

# Step 2: Impute missing values with the median
median_value = df['ConvertedCompYearly'].median()
df['ConvertedCompYearly'].fillna(median_value, inplace=True)

# Step 3: Check missing values after imputation
print("\nMissing Values After Imputation:")
print(df.isnull().sum())

# Step 4: Display updated DataFrame
print("\nUpdated DataFrame:")
print(df)
```

```python
# Impute missing values in categorical columns (e.g., `RemoteWork`) with the most frequent value.
# Step 1: Identify the most frequent (mode) value in 'RemoteWork'
mode_value = df['RemoteWork'].mode()[0]  # Get the most common value

# Step 2: Fill missing values with the mode
df['RemoteWork'].fillna(mode_value, inplace=True)

# Step 3: Verify missing values are filled
print("\nMissing Values After Imputation:")
print(df.isnull().sum())

# Step 4: Display updated DataFrame
print("\nUpdated DataFrame:")
print(df)
```

```python
#Apply Min-Max Scaling to normalize the `ConvertedCompYearly` column.
from sklearn.preprocessing import MinMaxScaler
```

```python
# Step 1: Initialize MinMaxScaler
scaler = MinMaxScaler()

# Step 2: Apply Min-Max Scaling to 'ConvertedCompYearly'
df['ConvertedCompYearly_MinMax'] = scaler.fit_transform(df[['ConvertedCompYearly']])

# Step 3: Display the updated DataFrame
print("\nNormalized Data:")
print(df)
```

```python
#Log-transform the ConvertedCompYearly column to reduce skewness.
import pandas as pd
import numpy as np
```

```python
# Step 1: Apply Log Transformation
df['ConvertedCompYearly_Log'] = np.log(df['ConvertedCompYearly'] + 1)

# Step 2: Display the transformed DataFrame
print("\nLog-Transformed Compensation:")
print(df[['ConvertedCompYearly', 'ConvertedCompYearly_Log']])
```
