```python
!wget -O survey-data.csv https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv

# Step 2: Import necessary libraries and load the dataset
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
df = pd.read_csv("survey-data.csv")

# Display the first few rows to understand the structure of the data
df.head()
```

```python
#1.1 Create a Pie Chart of the Top 5 Databases Respondents Want to Work With
df_db = df["DatabaseWantToWorkWith"].dropna()
df_db = df_db.str.split(";")
df_db = df_db.explode()

db_counts = df_db.value_counts().nlargest(5)

plt.figure(figsize=(8, 8))
plt.pie(db_counts, labels=db_counts.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.Pastel1.colors)
plt.title("Top 5 Databases Respondents Want to Work With", fontsize=14)
plt.axis("equal")  # Equal aspect ratio ensures pie is a circle.
plt.tight_layout()
plt.show()
```

```python
#1.2
df_dev = df['DevType'].dropna()
df_dev = df_dev.str.split(';')
df_dev = df_dev.explode()
top_dev_roles = df_dev.value_counts().nlargest(5)
plt.figure(figsize=(8, 8))
plt.pie(top_dev_roles, labels=top_dev_roles.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
plt.title("Top 5 Developer Roles (DevType)", fontsize=14)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.tight_layout()
plt.show()
```

```python
#1.3 Create a pie chart for the operating systems used by respondents for professional use
df_os = df['OpSysProfessional use'].dropna()
os_counts = df_os.value_counts()
plt.figure(figsize=(8, 8))
plt.pie(os_counts, labels=os_counts.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
plt.title("Distribution of Operating Systems Used Professionally", fontsize=14)
plt.axis('equal') 
plt.tight_layout()
plt.show()
```
