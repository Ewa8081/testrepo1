```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the Stack Overflow survey dataset
data_url = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv'
df = pd.read_csv(data_url)

# Display the first few rows of the dataset
df.head()
```

```python
print("Column Names:")
print(df.columns)
```

```python
print("\nData Types:")
print(df.dtypes)
```

```python
print("\nSummary Information:")
print(df.info())
```

```python
print("\nCategorical Data Summary:")
print(df.describe(include=['object']))
```

```python
missing_values = df.isna().sum()
print("Missing Values in Each Column:")
print(missing_values)
```

```python
years_mapping={"Less than 1 year":0.5,"More than 50 yers":51}
df["YearsCodePro"]=df["YearsCodePro"].replace(years_mapping)
```

```python
df["YearsCodePro"]=pd.to_numeric(df["YearsCodePro"],errors="coerce")
```

```python
df["YearsCodePro"].fillna(df["YearsCodePro"].median(),inplace=True)
```

```python
employment_counts = df["Employment"].value_counts()
print("Employment Distribution:\n", employment_counts)


jobsat_counts = df["JobSat"].value_counts()
print("\nJob Satisfaction Distribution:\n", jobsat_counts)


years_code_pro_counts = df["YearsCodePro"].value_counts()
print("\nYears of Professional Coding Experience Distribution:\n", years_code_pro_counts)
```

```python
plt.figure(figsize=(8, 6))
df["JobSat"].value_counts().plot.pie(autopct='%1.1f%%', startangle=90, cmap="viridis", wedgeprops={'edgecolor': 'black'})

plt.title("Job Satisfaction Distribution")
plt.ylabel("")
plt.show()
```

```python
def count_languages(column):
    language_series = df[column].dropna().str.split(';').explode()
    return language_series.value_counts()


worked_with_counts = count_languages("LanguageHaveWorkedWith")
want_to_work_with_counts = count_languages("LanguageWantToWorkWith")


lang_comparison = pd.DataFrame({
    "Worked With": worked_with_counts,
    "Want to Work With": want_to_work_with_counts
}).fillna(0)


top_languages = lang_comparison.head(10)

top_languages.plot(kind='bar', figsize=(12, 6), colormap="viridis")

plt.title("Comparison of Programming Languages (Worked With vs. Want to Work With)")
plt.xlabel("Programming Languages")
plt.ylabel("Frequency")
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

```python
data = {
    "Region": ["North America", "North America", "Europe", "Europe", "Asia", "Asia"],
    "RemoteWork": ["Fully Remote", "Hybrid", "Fully Remote", "In-Office", "Hybrid", "Fully Remote"],
    "Count": [500, 300, 450, 250, 600, 350]
}

df = pd.DataFrame(data)


plt.figure(figsize=(10, 6))
sns.barplot(x="Region", y="Count", hue="RemoteWork", data=df, palette="coolwarm")


plt.xlabel("Region")
plt.ylabel("Count of Responses")
plt.title("Distribution of Remote Work by Region")
plt.legend(title="Remote Work Type")
plt.xticks(rotation=45)
plt.show()
```
